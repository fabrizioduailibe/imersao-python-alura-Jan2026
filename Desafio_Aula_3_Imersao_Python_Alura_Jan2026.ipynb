```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
df.head()
```

```python
# Dicionário para renomear as colunas
colunas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'salario_usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'modo',
    'company_location': 'local_empresa',
    'company_size': 'tamanho_empresa'
}

# Mapeamento para os valores das categorias
mapeamento_experiencia = {
    'EN': 'Junior',
    'MI': 'Pleno',
    'SE': 'Senior',
    'EX': 'Executivo'
}

mapeamento_emprego = {
    'FT': 'Tempo Integral',
    'PT': 'Meio Período',
    'CT': 'Contrato',
    'FL': 'Freelance'
}

mapeamento_tamanho = {
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
}

# Aplicando as traduções
df_traduzido = df.rename(columns=colunas)
df_traduzido['senioridade'] = df_traduzido['senioridade'].map(mapeamento_experiencia)
df_traduzido['contrato'] = df_traduzido['contrato'].map(mapeamento_emprego)
df_traduzido['tamanho_empresa'] = df_traduzido['tamanho_empresa'].map(mapeamento_tamanho)

# Visualizando o resultado
df_traduzido.head()
```

```python
# Criando um mapeamento para os nomes dos cargos
mapeamento_cargos = {
    'Data Scientist': 'Cientista de Dados',
    'Data Engineer': 'Engenheiro de Dados',
    'Data Analyst': 'Analista de Dados',
    'Machine Learning Engineer': 'Engenheiro de Machine Learning',
    'Analytics Engineer': 'Engenheiro de Analytics',
    'Solutions Engineer': 'Engenheiro de Soluções',
    'Research Scientist': 'Cientista de Pesquisa',
    'AI Engineer': 'Engenheiro de IA',
    'Data Architect': 'Arquiteto de Dados',
    'Business Intelligence Analyst': 'Analista de Business Intelligence',
    'Data Science Manager': 'Gerente de Ciência de Dados',
    'Data Manager': 'Gerente de Dados',
    'ML Engineer': 'Engenheiro de ML',
    'Engineer': 'Engenheiro',
    'Data Specialist': 'Especialista de Dados',
    'Data Operations Manager': 'Gerente de Operações de Dados',
    'Data Science Consultant': 'Consultor de Ciência de Dados',
    'Data Science Architect': 'Arquiteto de Ciência de Dados',
    'Data Science Engineer': 'Engenheiro de Ciência de Dados',
    'Software Engineer': 'Engenheiro de Software',
    'Data Analytics Manager': 'Gerente de Análise de Dados',
    'Data Analytics Engineer': 'Engenheiro de Análise de Dados',
    'Product Manager': 'Gerente de Produtos'
}

# Aplicando a tradução na coluna 'cargo'
# Usamos replace para substituir apenas os valores mapeados, mantendo os outros como estão
df_traduzido['cargo'] = df_traduzido['cargo'].replace(mapeamento_cargos)

# Exibindo os primeiros registros para conferir
df_traduzido.head()
```

```python
# Checando se ainda sobraram registros com 'Data Scientist' escritos de outras formas
df_traduzido['cargo'].str.contains('Data Scientist', case=False).sum()
```

```python
# Checando visualmente os cargos desses 56 registros
df_traduzido[df_traduzido['cargo'].str.contains('Data Scientist', case=False)]
```

Como nenhum deles é puro "Data Scientist", não vamos considerá-los no desafio. Vamos considerar apenas cargos exatamente "Data Scientist", que foram todos traduzidos para "**Cientista de Dados**".

```python
# Removendo todos os campos nulos e aproveitando para ajustar os índices
df_limpo = df_traduzido.dropna().reset_index(drop=True)
df_limpo.info()
```

```python
# Confirmando se ainda há registros com campos nulos
df_limpo.isnull().sum()
```

```python
df_limpo.isnull().any(axis=1).sum()
```

```python
df_limpo.head()
```

```python
df_limpo[df_limpo['cargo'] == "Cientista de Dados"]
```

```python
# Obtendo apenas os registros de "Cientista de Dados"
df_limpo_cientdados = df_limpo[df_limpo['cargo'] == "Cientista de Dados"].reset_index(drop=True)
df_limpo_cientdados
```

```python
# Agruopando as medias salariais dos Cientistas de Dados por país, da maior para a menor.
media_salario_pais_cd = df_limpo_cientdados.groupby('local_empresa')['salario_usd'].mean().sort_values(ascending=False).reset_index()
media_salario_pais_cd
```

```python
# Plotando o gráfico de barras da Média Salarial Anual (USD) do cargo de Cientita de Dados por país, usando Plotly
import plotly.express as px
```

```python
fig = px.bar(
    media_salario_pais_cd,
    x='local_empresa',
    y='salario_usd',
    title='Média Salarial Anual do CIENTISTA DE DADOS por País - Plotly',
    labels={'local_empresa': 'País', 'salario_usd': 'Média Salarial Anual (USD)'},
    color='local_empresa',
    template='plotly_white',
    width=1200,
    height=700
)

fig.show()
```
