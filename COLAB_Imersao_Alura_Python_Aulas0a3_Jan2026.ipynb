### **Criando um dataframe do zero, para treinar possibilidades de substituição de campos nulos.**

```python
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ['João', 'Maria', 'José', 'Ana', 'Val'],
    'idade': [25, 30, 35, 40, 29],
    'salario': [5000, np.nan, 7000, np.nan, 23000]
})

# Preenchendo os campos Nan com a MÉDIA dos salários, arredondada paa 2 casas decimais.
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
print("Usando MÉDIA:")
print(df_salarios)


# Como temos um salário "outlier" (23000), o ideal é usar MEDIANA, para não preencher os campos
# nulos com valores muito destoantes da maioria.
df_salarios = pd.DataFrame({
    'nome': ['João', 'Maria', 'José', 'Ana', 'Val'],
    'idade': [25, 30, 35, 40, 29],
    'salario': [5000, np.nan, 7000, np.nan, 23000]
})

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
print("\nUsando MEDIANA:")
display(df_salarios)
```

### **OBS.:** Em alguns casos, nem a média nem a mediana podem ser adequadas. Pode ser melhor usar dados mais frequentes de outro período, ou simplesmente excluir as linhas. Tudo depende das especificidades da informação.

```python
# Testando um dataframe de temperaturas, pedindo ao Python para preencher automaticamente
# os valores nulos da coluna com o valor não nulo anterior -- ffill (forward fill)
df_temperaturas = pd.DataFrame({
    'dia_semana': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [np.nan, 30, np.nan, 28, np.nan]
})

# Incluindo uma coluna nova copiando a coluna 'temperaturas' e preenchendo os campos nulos
df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
print("Usando somente 'ffill':")
display(df_temperaturas)

# Pedindo ao Python para preencher automaticamente os valores nulos da coluna com o
# valor não nulo posterior -- bfill (backward fill)
df_temperaturas = pd.DataFrame({
    'dia_semana': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [np.nan, 30, np.nan, 28, np.nan]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].bfill()
print("\nUsando somente 'bfill':")
display(df_temperaturas)

# Como vimos, dependendo de não haver valor não nulo posterior ou anterior, ainda podem
# restar campos nulos mesmo depois da cada um destes métodos isoladamente. Então usa-se
# ambos, consecutivamente, para garantir máximo preenchimento de nulos.
df_temperaturas = pd.DataFrame({
    'dia_semana': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [np.nan, 30, np.nan, 28, np.nan]
})

df_temperaturas['preenchidos'] = df_temperaturas['temperatura'].ffill()
df_temperaturas['preenchidos'] = df_temperaturas['preenchidos'].bfill()
print("\nUsando 'ffill' e 'bfill':")
display(df_temperaturas)
```

```python
# Testando um dataframe de Cidades, substituindo campos nulos por um texto definido -- "Não informado"

df_cidades = pd.DataFrame({
    'nome': [ 'Ana', 'Fabrizio', 'Carlos', 'Val', 'João'],
    'cidade': ['São Luís', np.nan, 'Belém', np.nan, 'Souza']
})

# Incluindo uma coluna nova copiando a coluna de cidades e preenchendo os campos nulos
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
display(df_cidades)
```

```python
import pycountry
```

```python
# O Pycountry só trabalha com siglas de países de 3 letras.
# Função que converte as siglas do DataFrame de iso-2 para iso-3.
def iso2_para_iso3(code):
  try:
    return pycountry.countries.get(alpha_2=code).alpha_3
  except:
    return None

# Adiciona nova coluna ao DataFrame com os países em formato iso-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_para_iso3)
df_limpo['local_empresa_iso3'] = df_limpo['local_empresa'].apply(iso2_para_iso3)
df_limpo.head()
```

```python
# Cria dataframe apenas com os registros com cargo "Data Scientist"
df_ds = df_limpo[df_limpo['cargo'] == "Data Scientist"]
df_ds.head()
```

```python
# Cria o bloco de dados para o gráfico, agrupando apenas Pais de residência e Média salarial
media_sal_pais_ds = df_ds.groupby('residencia_iso3')['salario_usd'].mean().reset_index()
media_sal_pais_ds.head()
```

```python
import plotly.express as px
```

```python
# Gerar o gráfico lúdico (choropleth)
fig = px.choropleth(
    media_sal_pais_ds,
    locations='residencia_iso3',
    color='salario_usd',
    hover_name='residencia_iso3',
    title='Média Salarial Anual de Cientista de Dados por País',
    color_continuous_scale='rdylgn',     # 'rdylgn' (cores quentes), 'viridis' (cores frias)...
    labels={'salario_usd': 'Salário (USD)', 'residencia_iso3': 'País de Residência'},
    width=1000
)

fig.show()
```

```python
# Crinado o gráfico com base no país da empresa. É o mais coerente, pois o salário é diferente devido
# ao país da empresa empregadora e não devido ao país de residência do empregado.
media_sal_paisempresa_ds = df_ds.groupby('local_empresa_iso3')['salario_usd'].mean().reset_index()
media_sal_paisempresa_ds.head()
```

```python
# Gerar o gráfico lúdico (choropleth)
fig = px.choropleth(
    media_sal_paisempresa_ds,
    locations='local_empresa_iso3',
    color='salario_usd',
    hover_name='local_empresa_iso3',
    title='Média Salarial Anual de Cientista de Dados por País',
    color_continuous_scale='viridis',
    labels={'salario_usd': 'Salário (USD)', 'local_empresa_iso3': 'País da Empresa'},
    width=1000
)

fig.show()
```
