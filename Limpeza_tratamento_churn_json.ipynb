### Normalizando todas as coluna no objeto Json, ao invés do DataFrame

```python
import json
import pprint  # Pretty Print
```

```python
# Abrir arquivo json e carregar todo o conteúdo na variável
#    json_churn = json.load(open('dataset-telecom-churn-alura.json'))

# Ou podemo usar "with", que abre e fecha o arquivo, entregando o mesmo resultado:
with open('dataset-telecom-churn-alura.json') as json_file:
  json_bruto = json.load(json_file)
```

```python
pprint.pprint(json_bruto[:2])
```

```python
# Traduzindo todos os dados do Json para português brasileiro

# Dicionários de mapeamento para tradução
traducao_chaves = {
    'customerID': 'id_cliente', 'Churn': 'churn', 'customer': 'cliente', 'gender': 'genero',
    'SeniorCitizen': 'idoso', 'Partner': 'parceiro', 'Dependents': 'dependentes',
    'tenure': 'meses_contrato', 'phone': 'telefone', 'PhoneService': 'servico_telefone',
    'MultipleLines': 'multiplas_linhas', 'internet': 'internet', 'InternetService': 'servico_internet',
    'OnlineSecurity': 'seguranca_online', 'OnlineBackup': 'backup_online',
    'DeviceProtection': 'protecao_dispositivo', 'TechSupport': 'suporte_tecnico',
    'StreamingTV': 'streaming_tv', 'StreamingMovies': 'streaming_filmes',
    'account': 'conta', 'Contract': 'contrato', 'PaperlessBilling': 'fatura_digital',
    'PaymentMethod': 'metodo_pagamento', 'Charges': 'cobrancas', 'Monthly': 'mensal', 'Total': 'total'
}

traducao_valores = {
    'No': 'Não', 'Yes': 'Sim', 'Female': 'Feminino', 'Male': 'Masculino',
    'DSL': 'DSL', 'Fiber optic': 'Fibra', 'No phone service': 'Sem telefone',
    'No internet service': 'Sem internet', 'Month-to-month': 'Mensal',
    'One year': 'Um ano', 'Two year': 'Dois anos', 'Mailed check': 'Cheque Enviado',
    'Electronic check': 'Cheque Eletrônico', 'Credit card (automatic)': 'Cartao crédito (auto)',
    'Bank transfer (automatic)': 'Transferência (auto)'
}

def traduzir_e_formatar(obj):
    """Função recursiva para traduzir chaves/valores e aplicar formatação (lowercase e underscore)."""
    if isinstance(obj, dict):
        novo_obj = {}
        for chave, valor in obj.items():
            #print(f"Chave: {chave} | Valor: {valor}")
            # Traduz "chave" ou deixa a mesma "chave" se não encontrar no dicionário.
            # Força aplicação de minúsculas e substituição de espaços, tendo a "chave"
            # sido encontrada ou não no dicionário.
            nova_chave = traducao_chaves.get(chave, chave).lower().replace(' ', '_')
            # Submete o Valor a nova recursão, pois pode ser outro nó aninhado (dicionário),
            # lista, string ou número.
            novo_obj[nova_chave] = traduzir_e_formatar(valor)
        return novo_obj
    elif isinstance(obj, list):
        # Submete cada item da Lista a nova recursão, pois podem ser outros nós aninhados
        # (dicionários), listas, strings ou números.
        return [traduzir_e_formatar(item) for item in obj]
    # A partir daqui, chegou no último nível de aninhamento e só traduz (se for string)
    # e retorna os itens encontrados, sem fazer nova recursão.
    elif isinstance(obj, str):
        # Traduz valor string
        valor_traduzido = traducao_valores.get(obj, obj)
        return str(valor_traduzido)
    else:
        # Apenas retorna valor numérico
        return obj

# Aplicando a tradução no objeto json_bruto
json_traduzido = traduzir_e_formatar(json_bruto)

# Exibindo os dois primeiros registros traduzidos para verificação
pprint.pprint(json_traduzido[:2])
```

```python
# Normalizar (desaninhar) todas as colunas do Json de uma só vez e atribuir a um DataFrame
dados_normalizados = pd.json_normalize(json_traduzido)
dados_normalizados.head()
```
